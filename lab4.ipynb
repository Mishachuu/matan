```python
def equation1(x):
    return np.sin(x + 1) - 1.2

def equation2(y):
    return (1-np.cos(y - 2))/2

x_vals = np.linspace(-1, 5, 400)
y_vals = np.linspace(-1, 5, 400)

plt.plot(x_vals, equation1(x_vals), label='sin (x+1)−y-1.2=0')
plt.plot(equation2(y_vals), y_vals, label='cos(y)+2x-2=0')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.title('Графики функций')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def equations(vars):
    x, y = vars
    eq1 = np.sin(x + 1) - y - 1.2
    eq2 = 2*x + np.cos(y)-2
    return [eq1, eq2]

solution = fsolve(equations, (0, 0))
print("Решение системы уравнений:", solution)
```

```python
x, y = smp.symbols('x y')

F = smp.sin(x + 1) - y - 1.2
G = 2*x + smp.cos(y) - 2

f_dx = smp.diff(F, x)
f_dy = smp.diff(F, y)
g_dx = smp.diff(G, x)
g_dy = smp.diff(G, y)


J = smp.Matrix([[f_dx, f_dy], [g_dx, g_dy]])

print("Матрица Якоби:\n", J)

dx, dy = smp.symbols('dx dy')

linear_system = smp.Eq(J * smp.Matrix([dx, dy]), -smp.Matrix([F, G]))

print("Система линейных уравнений:\n")
print(linear_system)
```
