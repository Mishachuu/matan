Задание 6

```python
def y1(x):
    return 1+np.sin(x)


def y2(x):
    return (2+x)**0.5


def equation(x):
    return y1(x) - y2(x)


x_ = fsolve(equation, 0)[0]
y_ = y1(x_)
x_2 = fsolve(equation, np.pi)[0]
y_2 = y1(x_2)

x = np.linspace(-10, 10, 100)
plt.plot(x, y1(x), label='y1 = 1+sinX')
plt.plot(x, y2(x), label='y2 = sqrt(2+x)')

plt.plot(x_, y_, 'ro', label='Точка пересечения')
plt.plot(x_2, y_2, 'ro')


plt.legend(loc='upper left', frameon=False)
plt.grid()
plt.show()
```
